```python
#scalars and tensors

import tensorflow as tf
x=tf.constant(1)
print("scalar:",x)

a=tf.constant([1,2,3])
print("1d tensor:",a)
print("shape:",a.shape)

b=tf.constant([[1,2,3],[4,5,6],[7,8,9]])
print("2d tensor:\n",b)
print("shape:", tf.rank(b))
```

```python
#operations 

c=tf.constant([[1,2,3],[1,2,3],[1,2,3]])
print("addition:",tf.add(b,c))
print("multiplication:",tf.matmul(b,c))
print("element-wise multiplication", tf.multiply(b,c))
```

```python
#variables

v=tf.Variable(5)
print("initial:",v)
v.assign(v+1)
print("after+1:",v)
v.assign(20)
print("after assign:",v)
v.assign_add(10)
print("after assign add:",v)
v.assign_sub(5)
print("after assign sub:",v)
```

```python
#reshaping
c=tf.constant([1,2,3,4,5,6])
print("reshaped:\n",tf.reshape(c,(2,3)))

#type casting
d=tf.constant([1,2,3])
print("casting:\n",tf.cast(d,tf.float32))

#shape info of matrix
matrix=tf.constant([[1,2,3,],[2,3,4],[3,4,5]])
print("shape:",tf.shape(matrix))
print("rank:",tf.rank(matrix))
print("size:",tf.size(matrix))
```

```python
#expanded dimensions - makes 1d into 2d
v=tf.constant([1,2,3])
print("expanded with axis=0:\n",tf.expand_dims(v,axis=0)) #o/p is 1 row 3 cols
print("expanded with axis=1:\n",tf.expand_dims(v,axis=1)) #o/p is 3 rows 1 col
```

```python
#indexing and slicing

v1=tf.constant([1,2,3,4,5,6])
print("elem at idx 2:",v1[2].numpy())
print("slice 1-3",v1[1:3].numpy())
print("every 2nd elem:",v1[::2].numpy())
print("up to idx 3:",v1[:3].numpy())
print("from idx 2:",v1[2:].numpy())
print("----------------")
v2=tf.constant([[1,2,3],[4,5,6],[7,8,9]])
print("elem at (1,2):",v2[1,2].numpy())
print("row 0:",v2[0].numpy())
print("col 1:",v2[:,1].numpy())
print("submatrix\n",v2[0:2,1:3].numpy())
print("every 2nd row:\n",v2[::2,:].numpy())
```
